# Explorando a matriz de pixeis da imagem

Vimos que a imagem é composta por uma matriz de pixeis. A dimensão dessa matriz representa a dimensão da imagem (largura e altura).

Cada pixel é composto pela tripla RGB (R=Red, G=Green, B=Blue). E cada componente possui uma escala de 0 a 255.

**Desafio:** percorrer a matriz de pixels e "clarear" a imagem. Utilize pelo menos DUAS formas diferentes de clarear a imagem e verifique qual obteve o melhor resultado (ex: clarear todos os pixels ou somente aqueles pixels mais escuros). Verifique o resultado e explique qual abordagem, em sua opinião, apresentou melhores resultados de clareamento.

**Dicas e funções que podem ser utilizadas:**

img.load(): retorna uma matriz de pixels. Cada pixel da matriz possui uma tupla com os valores para os 3 canais de cores (R, G, B). 

img.size[0]: retorna a largura da imagem

img.size[1]: retorna a altura da imagem


**OBS 1:** não utilizar funções prontas. Execute o clareamento de forma manual passando por cada pixel da matriz.

**OBS 2:** A matriz de pixels retornada pela função load() é passada por referência. Cada vez que o clareamento for executado, a imagem ficará cada vez mais clara. Para evitar esse problema, uma solução é carregar a imagem no bloco de código de clareamento.

```python
from PIL import Image
import requests
from io import BytesIO
from IPython.display import display

url_imagem = 'https://dinaproductions.com/wp-content/uploads/2017/02/Aya-BE.jpg'

# lendo a imagem
response = requests.get(url_imagem)
img = Image.open(BytesIO(response.content))

px = img.load()

w = img.size[0]
h = img.size[1]

lightAll = Image.new("RGB", (w, h))

for x in range(w):
  for y in range(h):
    pxl = px[x, y]
    lightAll.putpixel((x, y), ((pxl[0]+10), (pxl[1]+10), (pxl[2]+10)))

display(lightAll)
```

## **ATIVIDADE 2**
Pense no sistema RGB, pesquise e verifique como deixar a imagem em **tons de cinza**. Tente encontrar duas formas diferentes de trabalhar com os pixels do sistema RGB para fazer essa transformação, verifique o resultado e explique qual abordagem, em sua opinião, apresentou uma melhor qualidade. Faça a transformação na imagem clareada com melhor resultado.

```python
from PIL import Image
import requests
from io import BytesIO
from IPython.display import display

url_imagem = 'https://dinaproductions.com/wp-content/uploads/2017/02/Aya-BE.jpg'

# lendo a imagem
response = requests.get(url_imagem)
img = Image.open(BytesIO(response.content))

px = img.load()

w = img.size[0]
h = img.size[1]

lightAll = Image.new("RGB", (w, h))
grey = Image.new("RGB", (w, h))

for x in range(w):
  for y in range(h):
    pxl = px[x, y]
    lightAll.putpixel((x, y), ((pxl[0]+10), (pxl[1]+10), (pxl[2]+10)))

px = lightAll.load()

for x in range(w):
  for y in range(h):
    pxl = px[x, y]
    med = (pxl[0] + pxl[1] + pxl[2])//3
    grey.putpixel((x, y), (med, med, med))

display(lightAll)
display(grey)
```
